```python
import matplotlib.pyplot as plt
l1=[]    #formation of empty list
b=[1,2,3,4]
for i in range(1,5):   #using of for loop for years 
    print(i,"year")
    x=int(input("enter total marks in mid term"))   #input of mid term marks
    y=int(input("enter total marks in term end"))   #input of term end marks
    p=int(input("enter internal marks"))            #input of internal marks
    m=(x*0.30)+(y*0.30)+(p*0.40)
    r=(m/294)*100
    l1.append(int(r)) #add elements in list
print(l1)
print("1-line graph")
print("2-bar graph")
print("3-piechart")
print("4-scatter plot")

a=int(input("enter your choice")) #choose your graph

if(a==1):
    plt.plot(b,l1)    #making of line graph
    plt.title("Line Graph")  #title of graph
    plt.xlabel("year")        #naming x-axis
    plt.ylabel("percentage")  #naming y-axis
    plt.legend()
    plt.show()   #display graph

elif(a==2):
    plt.bar(b,l1)    #making of bar graph
    plt.title("bar graph")   #title of graph
    plt.xlabel("year")       #naming x-axis
    plt.ylabel("percentage")  #naming y-axis
    plt.show()  #display graph

elif(a==3):
    plt.pie(l1,labels=b, autopct='%1.1f%%')
    plt.title("Pie Chart")  #title of graph
    plt.show()   #display graph

elif(a==4):
    plt.scatter(b,l1)
    plt.title("Scatter Plot")   #title of graph
    plt.xlabel("year")          #naming x-axis
    plt.ylabel("percentage")   #naming y-axis
    plt.show()   #display graph
else:    #else statement
    print("invalid choice")
```
